##Membuat dan Mengakses Elemen Series

```python
import pandas as pd

li = [1, 7, 2]
se = pd.Series(li)

print(se)

li = [1, 7, 2]
se = pd.Series(li)

print(se[0])

li = [1, 7, 2]
se = pd.Series(li, index = ['x', 'y', 'z'])

print(se['x'])

calories = {"day1": 420, "day2": 380, "day3": 390}
se = pd.Series(calories)

print(se)
```
